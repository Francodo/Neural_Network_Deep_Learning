```python
# (1) Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd
import tensorflow as tf
```

```python
# (2) Import our input dataset which is the charity_data.csv
charity_df = pd.read_csv('/users/Test/Resources/charity_data.csv')
charity_df.head()
```

```python
charity_df = charity_df.drop(columns="NAME")
```

```python
# (3) Generate our categorical variable list
charity_cat = charity_df.dtypes[charity_df.dtypes == "object"].index.tolist()

display(charity_cat)
```

```python
# Check the number of unique values in each column
charity_df[charity_cat].nunique()
```

```python
charity_df.head()
```

```python
# (4) Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# (5) Fit and transform the OneHotEncoder using the categorical variable list
charity_encode_df = pd.DataFrame(enc.fit_transform(charity_df[charity_cat]))

# (6) Add the encoded variable names to the DataFrame
charity_encode_df.columns = enc.get_feature_names(charity_cat)
charity_encode_df.head()
```

```python
# (7) Merge one-hot encoded features and drop the originals
charity_df = charity_df.merge(charity_encode_df,left_index=True, right_index=True)

charity_df = charity_df.drop(charity_cat,1)

charity_df.head()
```

```python
# (8) Split into features and target arrays. Remove or drop "Is_Successful" column as is the feature target
y = charity_df.IS_SUCCESSFUL.values
X=charity_df.drop(columns="IS_SUCCESSFUL").values


# (9) Split training/test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78, stratify=y)
```

```python
# (10) Create a StandardScaler instance
scaler = StandardScaler()

# (11) Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# (12) Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# (13) Define the model - deep neural network
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  6
hidden_nodes_layer2 = 4

nn = tf.keras.models.Sequential()

# (14) First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# (15) Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# (16) Check the structure of the model
nn.summary()
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Compile the Sequential model together and customize metrics
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)

# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
